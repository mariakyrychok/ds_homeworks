## Task about lakes


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(555)
a = np.random.randint(1,100,20)

plt.plot(a)
plt.show()

def get_depths(points):
    plt.plot(points)
    left = np.zeros(len(points))
    right = np.zeros(len(points))
    lakes_depth = np.zeros(len(points))
    
    left[0] = points[0]
    for idx in range(1, len(points)):
        left[idx] = max(left[idx-1], points[idx])
    
    right[-1] = points[-1]
    for idx in range(len(points)-2, -1, -1):
        right[idx] = max(right[idx+1], points[idx])
        
    for idx in range(len(points)):
        lakes_depth[idx] = min(left[idx], right[idx]) - points[idx]
    return lakes_depth

depths = get_depths(a)

left, lake_depth = 0, 0
deepest_right, deepest_left, deepest_lake_depth = 0, 0, 0
for right, depth in enumerate(depths):
    lake_depth = max(lake_depth, depth)
    if depth == 0:
        if lake_depth > deepest_lake_depth:
            deepest_left, deepest_right, deepest_lake_depth = left, right, lake_depth
        left = right
        lake_depth = 0
    
plt.plot(range(deepest_left, deepest_right+1), a[deepest_left:deepest_right+1], c='red')
plt.show()
```
